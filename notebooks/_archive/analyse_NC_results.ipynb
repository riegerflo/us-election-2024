```python
df2 = df[(df["Contest Name"] == "US PRESIDENT") | (df["Contest Name"] == "NC GOVERNOR")].groupby(["Precinct", "Contest Name"])[["Total Votes", "Election Day", "Early Voting", "Absentee by Mail"]].sum()
df2
```

```python
df2["Difference"] = df2.groupby("Precinct").diff()["Total Votes"]
df2["Bullet Ballots"] = df2["Difference"]/df2["Total Votes"]*100
df2
```

```python
df2.dropna().sort_values("Bullet Ballots", ascending=False)
```

```python
df2.dropna().reset_index().set_index("Precinct").sort_values("Bullet Ballots", ascending=False)["Bullet Ballots"].plot.bar()
```
